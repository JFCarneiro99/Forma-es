### 1. Usando o método de funções de passagem de argumentos arbitrários, crie um programa Python em que o utilizador:
#### - Introduza 5 marcas de carros e 5 cores correspondentes, criando dois dicionários
#### - De seguida, se o utilizador introduzir uma cor, o programa retorma a marca do carro correspondente . Ao introduzir uma marca de carro, o programa retorna a cor correspondente.
#### - Se existir cores ou marcas repetidas, o programa deverá apresentar todas as ocorrências
#### - Pode utilizar o numero de funções que desejar

```python
def criar_dicionarios():
    marcas = {}
    cores = {}
    for i in range(5):
        marca = input(f"Digite a marca do carro {i + 1}: ")
        cor = input(f"Digite a cor correspondente: ")
        if marca in marcas:
            marcas[marca].append(cor)
        else:
            marcas[marca] = [cor]
        if cor in cores:
            cores[cor].append(marca)
        else:
            cores[cor] = [marca]
    return marcas, cores

def encontrar_correspondencia(marcas, cores, consulta):
    if consulta in marcas:
        return f"A cor correspondente a {consulta} é: {', '.join(marcas[consulta])}"
    elif consulta in cores:
        return f"A(s) marca(s) correspondente(s) a {consulta} é(são): {', '.join(cores[consulta])}"
    else:
        return "Nenhuma correspondência encontrada."

def main():
    marcas, cores = criar_dicionarios()

    while True:
        opcao = input("Digite uma marca de carro ou cor (ou 'sair' para encerrar): ")
        if opcao.lower() == 'sair':
            break
        resultado = encontrar_correspondencia(marcas, cores, opcao)
        print(resultado)

if __name__ == "__main__":
    main()
```

### 2. Repita o programa de 1, com utilizando:

### 3. Usando uma função de qualquer método e a função lambda, efetue um calculo em que:
#### - No primeiro input o utilizador insere uma série de numeros separados por virgulas
#### - No segundo input insere o operador que pretende realizar em todos os numeros(soma, subtração, multiplicação e divisão)
#### - O programa retorna o resultado

```python
def calcular(operacao, numeros):
    if operacao == "soma":
        return sum(numeros)
    elif operacao == "subtracao":
        return numeros[0] - sum(numeros[1:])
    elif operacao == "multiplicacao":
        resultado = 1
        for num in numeros:
            resultado *= num
        return resultado
    elif operacao == "divisao":
        resultado = numeros[0]
        for num in numeros[1:]:
            resultado /= num
        return resultado
    else:
        return "Operador inválido."

operacoes = {
    "soma": lambda x: sum(x),
    "subtracao": lambda x: x[0] - sum(x[1:]),
    "multiplicacao": lambda x: x[0] * x[1],
    "divisao": lambda x: x[0] / x[1]
}

entrada_numeros = input("Digite uma série de números separados por vírgula: ")
numeros = [float(num) for num in entrada_numeros.split(',')]

operacao = input("Digite o operador (soma, subtracao, multiplicacao, divisao): ")

if operacao in operacoes:
    resultado = operacoes[operacao](numeros)
    print(f"Resultado: {resultado}")
else:
    print("Operador inválido.")
```

### 4. Utilizando a biblioteca NumPy, crie um array unidimensional de numeros aleatórios de 20 elementos, com números de 0 a 100. O programa retorna os números que são pares e respetivo índices.

```python
import numpy as np

numeros = np.random.randint(0, 101, 20)

pares = [(numero, indice) for indice, numero in enumerate(numeros) if numero % 2 == 0]

print("Números pares e seus índices:")
for numero, indice in pares:
    print(f"Índice {indice}: {numero}")
```

### 5. Realize novamente o exercício 4 para 3 arrays unidimensionais aleatórios. Desenhe um gráfico em que apresente os indices no eixo do x, e os números pares no eixo do y.
#### Nota: Para desenhar o gráfico, faça num ficheiro Python à parte e faça a importação deste no ficheiro principal

```python
import matplotlib.pyplot as plt

def array(indices, numeros_pares, titulo):
    plt.figure(figsize=(10, 6))
    plt.bar(indices, numeros_pares, color='blue')
    plt.xlabel('Índices')
    plt.ylabel('Números Pares')
    plt.title(titulo)
    plt.grid(True)
    plt.show()

import numpy as np
from matplotlib import pyplot as pyplot

numeros1 = np.random.randint(0, 101, 20)
numeros2 = np.random.randint(0, 101, 20)
numeros3 = np.random.randint(0, 101, 20)

pares1 = [(numero, indice) for indice, numero in enumerate(numeros1) if numero % 2 == 0]
pares2 = [(numero, indice) for indice, numero in enumerate(numeros2) if numero % 2 == 0]
pares3 = [(numero, indice) for indice, numero in enumerate(numeros3) if numero % 2 == 0]

pares1_numeros, pares1_indices = zip(*pares1)
pares2_numeros, pares2_indices = zip(*pares2)
pares3_numeros, pares3_indices = zip(*pares3)

array(pares1_indices, pares1_numeros, 'Array 1 - Números Pares')
array(pares2_indices, pares2_numeros, 'Array 2 - Números Pares')
array(pares3_indices, pares3_numeros, 'Array 3 - Números Pares')
```
